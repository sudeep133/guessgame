```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments
from torch.utils.data import Dataset
import torch

MODEL_NAME = "distilgpt2"
tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)

# Format data as prompt-response pairs
def format_example(item):
    return f"Question: {item['question']}\nAnswer: {item['answer']}{tokenizer.eos_token}"

class HistoryDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        self.examples = []
        for item in data:
            text = format_example(item)
            enc = tokenizer(text, truncation=True, max_length=max_length,
                          padding="max_length", return_tensors="pt")
            input_ids = enc["input_ids"].squeeze()
            self.examples.append({
                "input_ids": input_ids,
                "attention_mask": enc["attention_mask"].squeeze(),
                "labels": input_ids.clone()
            })
    def __len__(self): return len(self.examples)
    def __getitem__(self, i): return self.examples[i]

dataset = HistoryDataset(training_data, tokenizer)
print(f"Dataset ready: {len(dataset)} examples")
```

```python
training_args = TrainingArguments(
    output_dir="./history_model",
    num_train_epochs=10,          # More epochs = better memorization for small dataset
    per_device_train_batch_size=2,
    warmup_steps=10,
    logging_steps=10,
    save_steps=50,
    learning_rate=5e-5,
    fp16=True,                    # Use GPU half-precision for speed
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

trainer.train()
print("✅ Training complete!")
```

```python
# Save the fine-tuned model
model.save_pretrained("./history_model_final")
tokenizer.save_pretrained("./history_model_final")
print("Model saved locally in Colab")
```

### Test the Fine-Tuned Model

Now that the model is trained and saved, let's load it and see how it performs on a sample question.

```python
from transformers import pipeline

# Load the fine-tuned model and tokenizer
model_path = "./history_model_final"
finetuned_tokenizer = GPT2Tokenizer.from_pretrained(model_path)
finetuned_model = GPT2LMHeadModel.from_pretrained(model_path)

# Create a text generation pipeline
generator = pipeline(
    'text-generation',
    model=finetuned_model,
    tokenizer=finetuned_tokenizer
)

print("Model and tokenizer loaded, ready for generation!")
```

```python
# Example question to test the model
question = "Question: Who made the steam engine"

# Generate a response
response = generator(question, max_new_tokens=100, num_return_sequences=1)

# Extract and print the generated text
generated_text = response[0]['generated_text']

# The model will output the question followed by its generated answer
# We want to extract just the answer part
answer_prefix = "Answer: "
if answer_prefix in generated_text:
    generated_answer = generated_text.split(answer_prefix, 1)[1].strip()
    # Further clean up by removing the eos_token if present
    generated_answer = generated_answer.replace(finetuned_tokenizer.eos_token, "").strip()
else:
    generated_answer = generated_text # Fallback if prefix not found

print("Question:", question.replace('Question: ', ''))
print("\nGenerated Answer:", generated_answer)

# You can try other questions here as well:
# question_2 = "Question: How did railroads transform economies and societies in the 19th century?"
# response_2 = generator(question_2, max_new_tokens=100, num_return_sequences=1)
# generated_text_2 = response_2[0]['generated_text']
# if answer_prefix in generated_text_2:
#     generated_answer_2 = generated_text_2.split(answer_prefix, 1)[1].strip()
#     generated_answer_2 = generated_answer_2.replace(finetuned_tokenizer.eos_token, "").strip()
# else:
#     generated_answer_2 = generated_text_2
# print("\n---\nQuestion:", question_2.replace('Question: ', ''))
# print("\nGenerated Answer:", generated_answer_2)
```
